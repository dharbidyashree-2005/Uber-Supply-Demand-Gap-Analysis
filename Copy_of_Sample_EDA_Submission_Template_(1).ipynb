## 3. ***Data Wrangling***

### Data Wrangling Code

```python
df['Request timestamp'] = pd.to_datetime(
    df['Request timestamp'],
    format='mixed',
    dayfirst=True
)

df['Drop timestamp'] = pd.to_datetime(
    df['Drop timestamp'],
    format='mixed',
    dayfirst=True
)

df['Hour'] = df['Request timestamp'].dt.hour

def time_slot(x):
    if x >= 0 and x < 5:
        return 'Early Morning'
    elif x >= 5 and x < 10:
        return 'Morning'
    elif x >= 10 and x < 17:
        return 'Afternoon'
    elif x >= 17 and x < 21:
        return 'Evening'
    else:
        return 'Night'

df['Time Slot'] = df['Hour'].apply(time_slot)

df.head()
```

### What all manipulations have you done and insights you found?

During the data wrangling process, the Request Timestamp and Drop Timestamp columns were converted into datetime format to enable time-based analysis. A new column named Hour was created by extracting the hour from the request timestamp. Based on the request hour, a Time Slot column was generated to classify requests into Early Morning, Morning, Afternoon, Evening, and Night.

The analysis revealed that the dataset contains missing values in the Driver ID and Drop Timestamp columns, mainly due to ride cancellations and situations where no cabs were available. The newly created features help in understanding demand patterns across different time periods and identifying the supply-demand gap more effectively.
